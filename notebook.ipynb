```python
import pandas as pd
import numpy as np

np.random.seed(123)

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {
    12: "winter",
    1: "winter",
    2: "winter",
    3: "spring",
    4: "spring",
    5: "spring",
    6: "summer",
    7: "summer",
    8: "summer",
    9: "autumn",
    10: "autumn",
    11: "autumn",
}


# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df["season"] = df["timestamp"].dt.month.map(lambda x: month_to_season[x])
    return df


# Генерация данных
df = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
df.to_csv("temperature_data.csv", index=False)
```

Найдем посезонные данные и аномалии

```python
import time
from joblib import Parallel, delayed
import multiprocessing
import warnings

warnings.filterwarnings("ignore")


def analyze_only_pandas(df):
    start_time = time.time()

    # заранее посчитаем статистики для всех городов и сезонов
    rolling_mean = (
        df.groupby("city")["temperature"].rolling(window=30).mean().reset_index()
    )

    season_stats = (
        df.groupby(["city", "season"])["temperature"].agg(["mean", "std"]).reset_index()
    )

    def find_anomalies(group):
        city = group["city"].iloc[0]
        season = group["season"].iloc[0]
        stats = season_stats[
            (season_stats["city"] == city) & (season_stats["season"] == season)
        ].iloc[0]

        mean, std = stats["mean"], stats["std"]
        is_anomaly = (group["temperature"] < (mean - 2 * std)) | (
            group["temperature"] > (mean + 2 * std)
        )
        return is_anomaly

    anomalies = (
        df.groupby(["city", "season"]).apply(find_anomalies).reset_index(drop=True)
    )

    execution_time = time.time() - start_time
    return rolling_mean, season_stats, anomalies, execution_time


# функция для обработки одного города
def process_city_joblib(city_data):
    rolling_mean = city_data["temperature"].rolling(window=30).mean()
    season_stats = city_data.groupby("season")["temperature"].agg(["mean", "std"])

    anomalies = []
    for season in city_data["season"].unique():
        season_data = city_data[city_data["season"] == season]
        stats = season_stats.loc[season]
        is_anomaly = (
            season_data["temperature"] < (stats["mean"] - 2 * stats["std"])
        ) | (season_data["temperature"] > (stats["mean"] + 2 * stats["std"]))
        anomalies.extend(is_anomaly)

    return rolling_mean, season_stats, anomalies


def analyze_city_joblib(df):
    start_time = time.time()

    cities = df["city"].unique()
    city_data = [df[df["city"] == city] for city in cities]

    # большой overhead на создание процессов
    # n_jobs = multiprocessing.cpu_count()
    n_jobs = 3
    results = Parallel(n_jobs=n_jobs)(
        delayed(process_city_joblib)(data) for data in city_data
    )

    rolling_means = pd.concat([res[0] for res in results])
    season_stats = pd.concat([res[1] for res in results])
    # аномалии перемешаны, но мы все равно сагрегируем далее
    anomalies = pd.concat([pd.Series(res[2]) for res in results])

    execution_time = time.time() - start_time
    return rolling_means, season_stats, anomalies, execution_time
```

pandas работает быстро, даже нет смысла ускорять. Но первый запуск на joblib медленный, а следующий уже разогрет, поэтому лишь в пару раз медленней только pandas.

```python
rolling_mean1, season_stats1, anomalies1, time1 = analyze_only_pandas(df)
print(f"Time only pandas: {time1:.2f} seconds")
```

```python
rolling_mean2, season_stats2, anomalies2, time2 = analyze_city_joblib(df)
print(f"Time with joblib: {time2:.2f} seconds")
```

```python
rolling_mean2, season_stats2, anomalies2, time2 = analyze_city_joblib(df)
print(f"Time with joblib: {time2:.2f} seconds")
```
